```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('premier-league-matches.csv')
data.head()
```

```python
data.dtypes
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
 data['Home'] = data['Home'].astype('category').cat.codes
data['Away'] = data['Away'].astype('category').cat.codes
# data.head()
```

```python
data['FTR'] = data['FTR'].astype('category').cat.codes
```

```python
data.dtypes
```

```python
data
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf =  RandomForestClassifier(min_samples_split=10, n_estimators=50, random_state=1)
Train = data[(data['Date'] > '2022-01-01') & (data['Date'] < '2023-01-01')]
Test = data[data['Date'] > '2023-01-01']
Train
```

```python
Test
```

```python
X =['Home'	,'HomeGoals','AwayGoals','Away', 'Wk']
y = ['FTR']
```

```python
rf.fit(Train[X], Train[y])
```

```python
predictions = rf.predict(Test[X])
predictions
```

```python
Test[y]
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(Test[y], predictions)
acc
```
